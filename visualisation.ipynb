```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
originial_data = pd.read_csv('cleaning_data5.csv' , encoding='latin1')
data = pd.read_csv('cleaning_data6.csv')
```

```python
data.describe()
```

```python
originial_data.describe()
```

```python
import matplotlib.pyplot as plt

# Avant imputation: valeurs manquantes
before_imputation = originial_data['Salons'].dropna()

# Après imputation: remplacer les NaN par les valeurs imputées
after_imputation = data['Salons']

# Tracer les histogrammes
plt.figure(figsize=(12, 6))

# Histogramme avant imputation
plt.subplot(1, 2, 1)
plt.hist(before_imputation, bins=20, edgecolor='black')
plt.title('Distribution avant imputation')
plt.xlabel('Salons')
plt.ylabel('Fréquence')

# Histogramme après imputation
plt.subplot(1, 2, 2)
plt.hist(after_imputation, bins=20, edgecolor='black')
plt.title('Distribution après imputation')
plt.xlabel('Salons')
plt.ylabel('Fréquence')

plt.tight_layout()
plt.show()
```
